```python
import re
import pandas as pd
import json
from functools import reduce

_df = df[[x=="nl.insomnia247.nailbiter.eldrinkopubbot.eldrinko.ElDrinkoPubBot" for x in df["loggerName"]]].loc[:,["instant","message"]]
_df = _df[[x[:2] in ["im","om","es","ss"] for x in _df["message"]]]
prog = re.compile("(om|im|es|ss)\([0-9]+\):\s*(.*)")

data = [dict(tag=prog.match(r["message"]).group(1),value=prog.match(r["message"]).group(2)) for r in _df.to_dict(orient="records") if prog.match(r["message"]) is not None]
# print(pd.DataFrame(data).head(10))
def reducer(acc,val):
#     print(f"acc: {acc}")
    l,d = acc
    if val["tag"]=="ss":
        if d:
            l.append(d)
            d = {}
    
    d[val["tag"]] = val["value"]
    return (l,d)
        
l,d = reduce(reducer,data,([],{}))
if d is not None:
    l.append(d)
_df = pd.DataFrame(l)
testdata_df = _df
testdata_df
```

```python
import json
import pandas as pd

res = []
for r in testdata_df.to_dict(orient="records"):
    res_rec = {}
    for k in ["ss","es"]: 
        res_rec[k] = r[k]
    im = json.loads(r["im"])
    del im["beerlist"]
    im["left"]["value"] = im["left"]["msg"]
    if "username" not in im["userData"]:
        im["userData"]["username"] = "null"
    del im["left"]["msg"]
    res_rec["im"] = im
    res.append(res_rec)
#     print(r["om"])
    om = None
    if( not pd.isna(r["om"]) ):
        om = {}
        om["left"],om["right"] = tuple(json.loads(r["om"]))
    res_rec["om"] = om
s = json.dumps(res,indent=2)
with open("/tmp/583dc2818985012c246518a9.json","w") as f:
    f.write(s)
    pass
print(s)
```
