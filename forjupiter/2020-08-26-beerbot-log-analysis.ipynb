```python
import pandas as pd

df = pd.read_json("/Users/nailbiter/Downloads/ProtoElDrinkoPubBot_2020-08-26_17:15:19.json",lines=True,orient="records")
df
```

```python
import re
import pandas as pd
import json
from functools import reduce

_df = df[[x=="nl.insomnia247.nailbiter.eldrinkopubbot.eldrinko.ElDrinkoPubBot" for x in df["loggerName"]]].loc[:,["instant","message"]]
_df = _df[[x[:2] in ["im","om","es","ss"] for x in _df["message"]]]
prog = re.compile("(om|im|es|ss)\([0-9]+\):\s*(.*)")

data = [dict(tag=prog.match(r["message"]).group(1),value=prog.match(r["message"]).group(2)) for r in _df.to_dict(orient="records") if prog.match(r["message"]) is not None]
# print(pd.DataFrame(data).head(10))
def reducer(acc,val):
#     print(f"acc: {acc}")
    l,d = acc
    if val["tag"]=="ss":
        if d:
            l.append(d)
            d = {}
    
    d[val["tag"]] = val["value"]
    return (l,d)
        
l,d = reduce(reducer,data,([],{}))
if d is not None:
    l.append(d)
_df = pd.DataFrame(l)
testdata_df = _df
```

```python
import json

_records = testdata_df.to_dict(orient="records")[-4:]
_records = [{**r,**{k:json.loads(r[k]) for k in ["im","om"]}} for r in _records]
_records = [{**r,"im":{k:v for k,v in r["im"].items() if k!="beerlist"}} for r in _records]
_records = [{"1t":f"{r['ss']} => {r['es']}", "2im":r["im"], "3om":r["om"]} for r in _records]

print(json.dumps(_records,indent=2,sort_keys=True,ensure_ascii=False))
```
