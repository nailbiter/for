```python
import pymongo
from pymongo import MongoClient
import pandas as pd
from pytz import timezone
from bson.codec_options import CodecOptions
import matplotlib.pyplot as plt
%matplotlib inline

#global const's
WEEKS = 4
TIMESTAMPS_PER_HOUR = 2
#global var's
client = MongoClient('localhost', 27017)
coll = client.admin.passwords
doc = coll.find_one({"key":"MONGOMLAB"})
mongopass = doc["value"]
```

```python
client = MongoClient(f"mongodb://nailbiter:{mongopass}@ds149672.mlab.com:49672/logistics")
timecoll = client.logistics["alex.time"].with_options(codec_options=CodecOptions(tz_aware=True,tzinfo=pytz.timezone('Asia/Tokyo')))
docs = list(obj for obj in timecoll.find().sort("date",pymongo.DESCENDING).limit((7*(WEEKS+1)+2)*24*TIMESTAMPS_PER_HOUR))
```

```python
df = pd.DataFrame(docs)
df["week"] = df["date"].apply(lambda d:d.date().isocalendar()[1])
# df["weekday"] = df["date"].apply(lambda d:d.date().isocalendar()[2])
df["elapsed_hours"] = 1/TIMESTAMPS_PER_HOUR

last_week_idx = df["week"].max()
df = df[(last_week_idx > df["week"]) & (df["week"] >(last_week_idx-WEEKS-1))]

# df.groupby(["week","category"]).sum()
arr = dict()
cats = set()
for name,group in df.groupby(["week","category"]):
    (week,cat) = name
    if(week not in arr):
        arr[week] = dict()
    arr[week][cat] = group["elapsed_hours"].sum()
    arr[week]["week"] = week
    cats.add(cat)

# vals = arr.values()
# sorted(vals,key lambda d)
grouped_df = pd.DataFrame(arr.values())
grouped_df = grouped_df.fillna(0)
grouped_df = grouped_df.transpose()
grouped_df
```

```python
for i in range(WEEKS):
    plt.figure(i+1)
    grouped_df[i].plot.pie(figsize=(16,10),autopct="%.2f")
```
