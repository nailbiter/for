```python
import sympy
from itertools import permutations
res = []
for p in permutations(["a","b","c","d"],4):
    res.append([sympy.Symbol(s) for s in p])
def GE(A):
    rows,cols = A.shape
    n = cols
    for j in range(cols):
        if(A[j,j]==0):
            pivotRow = j+1
            while (pivotRow<n) :
                if (A[pivotRow,j] == 0):
                    pivotRow = pivotRow+1
                else:
                    break
            temp = A[pivotRow,:]
            A[pivotRow,:] = A[j,:]
            A[j,:] = temp;
        for  i in range(j+1,rows):
            r = A[i,j]/A[j,j]
            A[i,:] = A[i,:] - r*A[j,:]
    return A
#     end
# end
        
m = GE(sympy.Matrix(res))
# m[0,0] = 0
# m.shape
m
```
