```python
import sympy
from itertools import permutations
res = []
for p in permutations(map(lambda x:sympy.Symbol(x),["a","b","c","d"]),4):
    res.append(list(p))
def GE(A,stop=-1):
    rows,cols = A.shape
    for j in range(cols):
        if(stop<0 or j<stop):
            if(A[j,j]==0):
                pivotRow = j+1
                while (pivotRow<rows) :
                    if (A[pivotRow,j] == 0):
                        pivotRow = pivotRow+1
                    else:
                        break
                for k in range(cols):
                    temp = A[pivotRow,k]
                    A[pivotRow,k] = A[j,k]
                    A[j,k] = temp
            for i in range(j+1,rows):
                r = A[i,j]/A[j,j]
                for k in range(cols):
                    A[i,k] = A[i,k] - r*A[j,k]
    return A
        
GE(sympy.Matrix(res))
```
