```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({"x": np.linspace(0, 2 * np.pi, 10000)})
df["y1"] = np.sin(df["x"])
df["y2"] = 10 * np.cos(df["x"])
df["y3"] = 1000 * df["x"]
df
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.lineplot(df, x="x", ax=ax, y="y1")
ax2 = ax.twinx()
sns.lineplot(df, x="x", ax=ax2, y="y2", color="orange")
ax3 = ax.twinx()
sns.lineplot(df, x="x", ax=ax2, y="y3", color="orange")
ax3.spines.right.set_position(("axes", 1.2))
```
