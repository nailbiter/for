```python
import pandas as pd
from os.path import splitext

df = pd.read_csv("/Users/nailbiter/Downloads/filelist.txt",sep="\t",header=None)
df
```

```python
df["ext"] = [splitext(fn)[1] for fn in df[0]]
df
```

```python
from itertools import chain
extensions_df = pd.read_csv("/Users/nailbiter/Downloads/audio_formats.csv",header=None)
extensions = pd.DataFrame({"ext":chain(*[[ee.strip() for ee in e.split(",")] for e in extensions_df[0]])}).set_index("ext")

music_df = pd.DataFrame({"path":extensions.join(df.set_index("ext"))[0]})
music_df = music_df.dropna()
music_df
```

```python
with open("/Users/nailbiter/Downloads/mp3_tag2.sh","w") as f:
    n = len(music_df["path"])
    for i,p in enumerate(music_df["path"]):
        f.write(f"echo \"{p}\"\n")
        f.write(f">&2 date\n")
        f.write(f">&2 echo \"{i+1}/{n}\"\n")
        f.write(f"eyeD3 \"{p}\"\n")
```
