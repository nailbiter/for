```python
from sympy import diff, Matrix, simplify
from sympy.abc import u,v,w

# def det(m):
def jacobian(new_vars,old_vars):
    m = Matrix([[simplify(diff(nv,ov)) for nv in new_vars] for ov in old_vars])
    return m.det()

x,y,z = u+v*w, v+w*u, w+u*v
simplify(jacobian([x,y,z],[u,v,w]))
```

```python
# 8

from sympy.abc import u,v
from random import uniform
import matplotlib.pyplot as plt

def transformation_plot(new_vars,old_vars,bounds,N=5000):
    assert(len(old_vars)==len(bounds))
    assert(len(new_vars)==2)
    nvs = []
    for i in range(N):
        old_var_sample = [uniform(b[0],b[1]) for b in bounds]
        new_var_point = [nv.subs(zip(old_vars,old_var_sample)) for nv in new_vars]
        nvs.append(new_var_point)
    return plt.scatter([n[0] for n in nvs],[n[1] for n in nvs],marker=".")

x = v
y = u*(1+v**2)
transformation_plot([x,y],[u,v],[[0,1],[0,1]])
```
